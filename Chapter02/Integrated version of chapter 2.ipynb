```python
#Examples of only importing a few function from a module
from scipy import log,exp,sqrt,stats
#BSM model appplication
def bscall(S,X,T,r,sigma):
    d1=(log(S/T)+(r+sigma*sigma/2.)*T)/(sigma*sqrt(T))
    d2=d1-sigma*sqrt(T)
    return S*stats.norm.cdf(d1)-X*exp(-r*T)*stats.norm.cdf(d2)

#Apply our defined function
bscall(40,40,0.1,0.05,0.2)
```
